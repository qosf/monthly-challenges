# Background
The k-NN algorithm is one of the best known classification algorithms and an example of supervised learning. The objective of this classification is that from a training set that are labeled by a class, for example the class label "1", "2", "dog", "cat", etc. we must from a test set identify from the algorithm to which class they belong.  The algorithm to be used is called k-NN (k-Nearest Neighbors), where k is the number of nearest neighbors that we are going to compare in order to choose which class it belongs to, therefore this classifier will consider the test set that we will call states |ai> whose labels are determined by finding the minimum distance to the states of the training set |bj> , whose labels are known beforehand [[1]](https://arxiv.org/pdf/2003.09187.pdf).

To find the distance between two instances, in classical computation one of these is the Euclidean distance, $\lvert a-b \rvert = \sqrt{\sum_{i=1}^N (a_i-b_i)^2}$. In the case of quantum computation, its equivalence is achieved from the inner product (or dot product) which is defined as $\lvert u-v \rvert = \lvert a \rvert  \lvert b \rvert -a \cdot b$, where |a> and |b> are two quantum states.To implement the inner product corresponding to the Euclidean distance, it is necessary to use a quantum subroutine known as the SWAP test [[2]](https://arxiv.org/pdf/1401.2142.pdf), which is a quantum algorithm that can be used to estimate the fidelity of two pure states $|a>$ and $|v>$, i.e., $F=\lvert <a|b>\rvert ^2$. Figure (1) shows the quantum circuit to perform such an operation
